## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

**Pertanyaan 1:** Seberapa besar perbedaan penurunan rata-rata jumlah penyewaan sepeda antara hari kerja dan hari libur saat cuaca hujan pada masing-masing musim (season) selama periode 2011-2012, dan strategi operasional apa yang dapat diterapkan untuk mengurangi dampak penurunan tersebut di setiap musim?

Keterangan:
  - **Specific**: Fokus pada "penurunan rata-rata jumlah penyewaan" yang membandingkan "hari kerja vs hari libur" dalam kondisi "cuaca hujan" di setiap "musim"
  - **Measurable**: Dapat diukur dengan menghitung rata-rata jumlah penyewaan dan mencari angka selisih penurunan yang terjadi
  - **Action-Oriented**: Hasil analisis dapat digunakan untuk strategi operasional, seperti penyesuaian distribusi speeda, atau pengelolaan kapasitas di tiap musim
  - **Relevant**:Relevan karena berkaitan langsung dengan penurunan penggunaan layanan yang berdampak pada performa bisnis bike sharing
  - **Time-bound**: Dibatasi pada periode tahun 2011-2012, sehingga analisis memiliki batas waktu yang jelas.

### Cleaning Data

#### Fixing ... problem

```python
#Mengubah kolom tanggal (dteday) pada dataset harian menjadi tipe datetime
day_df['dteday'] = pd.to_datetime(day_df['dteday'])

day_df.info()
```

```python
#Mengubah kolom tanggal (dteday) pada dataset per jam menjadi datetime
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])

hour_df.info()
```

```python
#Melakukan mapping data kategorikal (season,weather, weekday, workingday,year)
season_map = {
    1:'Winter',
    2:'Spring',
    3:'Summer',
    4:'Fall'
}

weather_map = {
    1:'Clear',
    2:'Cloudy',
    3:'Light Rain',
    4:'Heavy Rain'
}

weekday_map = {
    0:'Sunday',
    1:'Monday',
    2:'Tuesday',
    3:'Wednesday',
    4:'Thursday',
    5:'Friday',
    6:'Saturday'
}

workingday_map = {
    0:'Weekend',
    1:'Working day'
}

year_map = {
    0:'2011',
    1:'2012'
}

day_df['season'] = day_df['season'].map(season_map)
day_df['weathersit'] = day_df['weathersit'].map(weather_map)
day_df['weekday'] = day_df['weekday'].map(weekday_map)
day_df['workingday'] = day_df['workingday'].map(workingday_map)
day_df['yr'] = day_df['yr'].map(year_map)

hour_df['season'] = hour_df['season'].map(season_map)
hour_df['weathersit'] = hour_df['weathersit'].map(weather_map)
hour_df['weekday'] = hour_df['weekday'].map(weekday_map)
hour_df['workingday'] = hour_df['workingday'].map(workingday_map)
hour_df['yr'] = hour_df['yr'].map(year_map)

```

```python
#Visualisasi perbandingan penyewaan (casual vs registered) per jam

hourly_user_avg = hour_df.groupby('hr').agg({
    'casual':'mean',
    'registered':'mean'
}).reset_index()

x = np.arange(len(hourly_user_avg['hr']))
width = 0.35

plt.figure(figsize=(15,6))

plt.bar(x- width/2, hourly_user_avg['registered'],width,label='Registered',color='#72BCD4')
plt.bar(x+ width/2, hourly_user_avg['casual'],width,label='Casual',color='#FFA07A')

plt.title('Perbandingan Rata-rata Penyewa per Jam: Casual vs Registered',fontsize=18)
plt.xlabel('Jam (00-23)', fontsize=12)
plt.ylabel('Rata-rata Jumlah Penyewa', fontsize=12)
plt.xticks(hourly_user_avg['hr'])
plt.legend()
plt.grid(axis='y',linestyle='--',alpha=0.4)

plt.show()
```

```python
#Visualisasi pola penyewaan per jam antara hari kerja dan hari libur

hourly_avg_for_pivot = hour_df.groupby(['hr','workingday']).agg(
    mean_cnt = ('cnt','mean')
).reset_index()

pivot_hour = hourly_avg_for_pivot.pivot(index='hr',columns='workingday',values='mean_cnt')
plt.figure(figsize=(12,6))

plt.plot(pivot_hour.index,pivot_hour['Working day'], marker='o',label='Working Day')
plt.plot(pivot_hour.index, pivot_hour['Weekend'], marker='o', label='Holiday')

plt.title('Pola Penyewaan Sepeda Per Jam (Working Day vs Holiday)')
plt.xlabel('Jam')
plt.ylabel('Rata-rata cnt')
plt.xticks(range(24))
plt.legend()
plt.grid(True)

plt.show()
```

## Analisis Lanjutan (Opsional)

```python
#Membuat fungsi untuk mengelompokkan jam ke dalam kategori aktivitas (Morning Rush, Evening Rush, Business Hours, Night Leisure, Deep rest)
def mobility_segment(hr):
  if 7 <= hr <=9:
    return 'Morning Rush' #Jam berangkat kerja/sekolah
  elif 16 <= hr <= 19:
    return 'Evening Rush' #Jam pulang kerja/sekolah
  elif 10 <= hr <= 15:
    return 'Business Hours' #Jam aktif operasional
  elif 20 <= hr <= 23:
    return 'Night Leisure' #Jam ..
  else:
    return 'Deep Rest' #Jam istirahat

hour_df['mobility_type'] = hour_df['hr'].apply(mobility_segment)

print(hour_df['mobility_type'].unique())
```

```python
#Menghitung rata-rata penyewaan berdasarkan segmen waktu (mobility_type_
mobility_analysis = hour_df.groupby('mobility_type')['cnt'].mean().reset_index()
mobility_analysis
```

```python
#Menganalisis perbedaan pola penyewaan berdasarkan segmen waktu dan tipe hari

mobility_daytype = hour_df.groupby(['mobility_type','workingday'])['cnt'].mean()

mobility_daytype
```

```python
#Visualisasi hasil segmentasi (cluster) berdasarkan pola mobilitas
plt.figure(figsize=(10,6))

segmen_order = ['Morning Rush','Business Hours','Evening Rush', 'Night Leisure','Deep Rest']

warna_segmentasi = {
    'Morning Rush': '#4A90E2',
    'Business Hours': '#2193B0',
    'Evening Rush': '#0D6B7C',
    'Night Leisure': '#FF8C42',
    'Deep Rest': '#A8D8EA'
}

colors_bar = [warna_segmentasi [segmen] for segmen in segmen_order]
bars = plt.barh(segmen_order,mobility_analysis.set_index('mobility_type').loc[segmen_order,'cnt']
                ,color= colors_bar, edgecolor='white',linewidth=1.5)

plt.title('Rata-rata Penyewaan Sepeda Berdasarkan Segmen Mobilitas')
plt.xlabel('Segmen Waktu')
plt.ylabel('Rata-rata Jumlah Penyewaan')
plt.show()
```
